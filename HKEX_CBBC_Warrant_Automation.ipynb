```python
import pandas as pd 
import yfinance as yf
```

```python
#Variables to generate hkex websites 
num_all= ["%.2d" % i for i in range(1,13)]
```

Retrieve HKEX Callable Bull/Bear Contracts(CBBC) data 

```python
#Read zip csv with utf-16 encoding and tab as spacing 
for i in range(len(num_all)):
    url = 'https://www.hkex.com.hk/eng/cbbc/download/CBBC{}.zip'.format(num_all[i])
    globals()[f"df_cbbc_all_{i}"]= pd.read_csv(url, compression= 'zip',encoding='utf-16', sep='\t')
    print(url)
```

```python
#Data Cleansing 
for i in range(len(num_all)):
    globals()[f"df_cbbc_all_{i}_new"]= globals()[f"df_cbbc_all_{i}"]
    globals()[f"df_cbbc_all_{i}_new"]= globals()[f"df_cbbc_all_{i}"].drop(globals()[f"df_cbbc_all_{i}"].tail(4).index)
```

```python
#Combine all cbbc dataframe into one 
cbbc_list = [df_cbbc_all_0_new,df_cbbc_all_1_new,df_cbbc_all_2_new,df_cbbc_all_3_new,df_cbbc_all_4_new,df_cbbc_all_5_new,df_cbbc_all_6_new,df_cbbc_all_7_new,df_cbbc_all_8_new,df_cbbc_all_9_new,df_cbbc_all_10_new,df_cbbc_all_11_new]
df_cbbc= pd.concat(cbbc_list)
```

CBBC Data 

```python
df_cbbc
```

```python
#Format conversion from HKEX supported format to Yahoo Finance API 
df_cbbc['Underlying_Code']= df_cbbc['Underlying']  
name= '.HK'
df_name= df_cbbc[['Underlying','Underlying_Code']]
df_name1= df_name.drop_duplicates(subset=['Underlying'])
df_index= df_name1[(df_name1['Underlying'] == 'HSI')|(df_name1['Underlying'] == 'HSTEC')|(df_name1['Underlying'] == 'HSCEI')]
df_name2= df_name1.drop(df_name1[(df_name1['Underlying'] == 'HSI')|(df_name1['Underlying'] == 'HSTEC')|(df_name1['Underlying'] == 'HSCEI')].index)
df_name2['Underlying']= df_name2['Underlying']+ name
df_name2['Underlying']= df_name2['Underlying'].str[1:9]
#df_index2= df_index.replace({'Underlying' : { 'HSTEC' : '3032.HK', 'HSCEI' : '^HSCE'}})
df_index2= df_index.replace({'Underlying' : {'HSCEI' : '^HSCE'}})
df_name_final= pd.concat([df_name2, df_index2])
df_name_final
```

```python
#Retrieve data from Yahoo Finance API
db= pd.DataFrame()
for i in df_name_final['Underlying']: 
    data = yf.download(tickers = i,period = "2y",interval = "1d",)
    data['Stock Code']= i 
    db= pd.concat([db,data], axis=0)
```

Stock Price Performance(2Y) of listed CBBC 

```python
stock_db= db.reset_index()
stock_db_1= pd.merge(stock_db,df_name_final,how='left',left_on='Stock Code',right_on='Underlying').drop(columns=['Underlying'])
stock_db_1
```

```python
#Retrieve HKEX Warrant Data
for i in range(len(num_all)):
    url = 'https://www.hkex.com.hk/eng/dwrc/download/DW{}.zip'.format(num_all[i])
    globals()[f"df_warrant_all_{i}"]= pd.read_csv(url, compression= 'zip',encoding='utf-16', sep='\t')
    print(url)
```

```python
#Data Cleansing 
for i in range(len(num_all)):
    globals()[f"df_warrant_all_{i}_new"]= globals()[f"df_warrant_all_{i}"]
    globals()[f"df_warrant_all_{i}_new"]= globals()[f"df_warrant_all_{i}"].drop(globals()[f"df_warrant_all_{i}"].tail(4).index)
```

```python
#Combine all warrant list into one 
warrant_list = [df_warrant_all_0_new,df_warrant_all_1_new,df_warrant_all_2_new,df_warrant_all_3_new,df_warrant_all_4_new,df_warrant_all_5_new,df_warrant_all_6_new,df_warrant_all_7_new,df_warrant_all_8_new,df_warrant_all_9_new,df_warrant_all_10_new,df_warrant_all_11_new]
df_warrant= pd.concat(warrant_list)
```
